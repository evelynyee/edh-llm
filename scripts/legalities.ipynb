```python
import pandas as pd
import pickle
from scrape_cardlists import DATA_PATH
import os
CARDS_PATH = os.path.join(DATA_PATH, "cards_unique.pkl")
def load_data(fp):
    with open(fp, "rb") as f:
        data = pickle.load(f)
    return data
```

```python
cards = load_data(CARDS_PATH)
cards
```
